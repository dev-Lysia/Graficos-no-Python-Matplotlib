## Importação

```python
import matplotlib.pyplot as plt
```

### plot()
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html?highlight=pyplot%20plot#matplotlib-pyplot-plot

```python
x = [1, 2, 3, 4]
y = [2, 3, 4, 3]

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, label='dados')

plt.ylabel('Eixo Y') # nome da linha y
plt.xlabel('Eixo X') # nome da linha x

plt.title('Novo Gráfico')

plt.xticks([0,1,2,3,4,5,6]) # numeros das linha x
plt.yticks([0,2,4,6]) # numeros das linha y
plt.legend()

plt.show()
```

## LINES propriedades
https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_marker

```python
plt.plot(x, y, linestyle='--',color='r',lw=4.0)
plt.show()
```

## .Scatter()

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib-pyplot-scatter

```python
plt.scatter(x, y, color = 'r')
plt.show()
```

## .bar()

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib-pyplot-bar

```python
plt.bar(x,y, color='g')
plt.show()
```

### SUBPLOTS
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html


```python
valoresx = [1, 2, 3, 4]
valoresy = [1, 4, 2, 3]


figura= plt.figure(figsize=(40,15))
figura.suptitle('Título Geral') #titulo no inicio
facecolor='blue'

figura.add_subplot(131) # posição dos graficos
plt.plot(valoresx, valoresy,  label = ' Um dado qualquer')
plt.legend()
plt.title ('Gráfico 1')

figura.add_subplot(133)
plt.scatter(valoresx,valoresy)
plt.title ('Gráfico 2')

figura.add_subplot(132)
plt.title ('Gráfico 3')
plt.bar(valoresx, valoresy)

plt.savefig('grafico.png', dpi =100)
plt.show(figura)
```

## Aplicação: Fonte de dados Kaggle

https://www.kaggle.com/mattiuzc/stock-exchange-data

```python
import pandas as pd

cotacaodf = pd.read_csv('indexdata.csv')
display(cotacaodf)
```

```python
cotacaodf = cotacaodf[['Index','Date','Close']] #trazendo somente 3 colunas

display(cotacaodf)
```

```python
dfremove = cotacaodf.loc[(cotacaodf['Index'] != 'NYA')] #tirando tudo que tem diferente de index 'nya'
cotacao_final = cotacaodf.drop(dfremove.index) #deletando tudo que tem diferente de index 'nya'

cotacao_final = cotacao_final[13900:] #diminuindo a quantidade de linhas da tabela
display(cotacao_final)
```

```python
plt.figure(figsize=(20,10)) # tamanho da figura
plt.plot(cotacao_final['Date'], cotacao_final['Close'], label = 'NYA', color = 'g') # grafico(eixo x, eixo y)
plt.legend(loc=2, fontsize = 18)
plt.ylabel('Valor Fechamento')
plt.xlabel('Data da Cotacão')
plt.title('Grafico de cotacão historica da Bolsa NYA')
plt.xticks(['2021-03-23','2021-04-23','2021-05-03'])
plt.show()
```
